```python
# Import dependencies
import pandas as pd
```

```python
# Read the data
bauer_df = pd.read_csv('Data/bauer.csv')
bieber_df = pd.read_csv('Data/bieber.csv')
buehler_df = pd.read_csv('Data/buehler.csv')
cole_df = pd.read_csv('Data/cole.csv')
degrom_df = pd.read_csv('Data/degrom.csv')
kershaw_df = pd.read_csv('Data/kershaw.csv')
lamet_df = pd.read_csv('Data/lamet.csv')
nola_df = pd.read_csv('Data/nola.csv')
scherzer_df = pd.read_csv('Data/scherzer.csv')
```

```python
pitcher_df = pd.concat([bauer_df, bieber_df, buehler_df, cole_df, degrom_df, kershaw_df, lamet_df, nola_df, scherzer_df])
```

```python
# Save needed columns 
pitch_df = pitcher_df[['pitch_type', 'release_speed', 'player_name', 'zone', 'game_type', 'stand', 'p_throws', 
                       'type', 'balls', 'strikes', 'inning', 'effective_speed', 'release_spin_rate', 
                       'release_extension', 'release_pos_y', 'pitch_number', 'pitch_name']]
```
