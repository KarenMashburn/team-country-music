```python
# Import dependencies
import pandas as pd
```

```python
# Read in data
bb_2020_df = pd.read_csv('Data/Statcast/Statcast_2020.csv')
bb_2019_df = pd.read_csv('Data/Statcast/Statcast_2019.csv')
bb_2018_df = pd.read_csv('Data/Statcast/Statcast_2018.csv')
bb_2017_df = pd.read_csv('Data/Statcast/Statcast_2017.csv')
```

```python
baseball_df = pd.concat([bb_2020_df, bb_2019_df, bb_2018_df, bb_2017_df])
```

```python
# Convert 'game_date' to datetime 
baseball_df['game_date'] = pd.to_datetime(baseball_df['game_date'])
```

```python
# Save necessary columns to new df
pitch_df = baseball_df[['game_date', 'type', 'pitch_type', 'release_speed', 'effective_speed', 'release_spin_rate', 'release_pos_x', 'release_pos_z', 'zone', 'stand', 'p_throws', 'pitch_number', 'if_fielding_alignment', 'of_fielding_alignment', 'spin_axis']]
```

```python
# Check new df
pitch_df.head(10)
```

```python
# Drop columns with NAs 
pitch_df = pitch_df.dropna()
```

```python
# Export new df as csv file
pitch_df.to_csv('Data/Saved/pitching_all.csv')
```
